```python
# Finding the Highest Defense pokemon for electric types 
```

```python
def get_top_type_pokemon(df, top_n, type_name, select_stat_name):
    type_df = df[df['Type 1'] == type_name]

    sorted_type = type_df.sort_values(by=select_stat_name, ascending=False)

    top_type = sorted_type.head(top_n)

    return top_type
```

```python
top_electric_pokemon = get_top_type_pokemon(df, top_n=10, type_name='Electric', select_stat_name='Speed')
```

```python
#below i did an y='Name' and realised it changed the order of the graph compared to using hue but its recommended
#to use hue due to the fact that it will be removed later in the future. 
```

```python

plt.figure(figsize=(16,8))
sns.barplot(x='Speed', y='Name', data=top_electric_pokemon, palette='cividis')
plt.title=f"Top {top_n} Highest Speed Electric Pokemon"
plt.xlabel=('Speed')
plt.ylabel=('Name')
plt.show()
```
